# read cytokine level from each patient

```python
import pandas as pd
dfCytokine = pd.read_excel("Sepsis_patient_cytokine_levels.xlsx")
```

```python
dfCytokine.head()
```

# remove NaN value in dataframe

```python
df= dfCytokine.dropna(thresh=2)
```

```python
df.info()
```

# save into csv file

```python
df.to_csv("Sepsis_patient_cytokine_levels.csv")
```

# calculate mean of each cytokine level

```python
Mean = df.mean(axis=0)
```

```python
Mean
```

# add each mean to the last row of dataframe

```python
dfMeans = df.append(Mean, ignore_index=True)
```

```python
dfMeans
```

# save into csv file

```python
dfMeans.to_csv("Sepsis_patient_cytokine_levels_Mean.csv")
```
