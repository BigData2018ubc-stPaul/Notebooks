```r
## For cytokines with p < 0.05 in the dataset
## Correlating each of these cytokines to IL8
```

```r
install.packages("ggpubr")
```

```r
df = read.csv("Sepsis_patient_cytokine_levels.csv",header=T)
```

```r
df
```

```r
library("ggpubr")
ggscatter(df, x = "IL8", y = "EOTAXIN", 
          add = "reg.line", conf.int = TRUE, 
          cor.coef = TRUE, cor.method = "spearman",
          xlab = "IL8[conc]", ylab = "EOTAXIN[conc]")
```
