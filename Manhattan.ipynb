# GWAS - survival

```r
data<-read.table(file="as1.assoc", header=T)
head(data)
```

#install.packages('qqman')

```r
library('qqman')
```

```r
data<-na.omit(data)
png("GWAS-survival.png",width=1366,height=768,res=150)
manhattan(data, main = "GWAS-survivial", ylim = c(0, 7), cex = 0.6, 
         cex.axis = 1, cex.lab = 1, cex.main = 1, col = c("blue4", "red3"), genomewideline = F,
         annotatePval = 0.00001, annotateTop = FALSE)
dev.off()
```

```r
png("GWAS-survival-chr9.png",width=1366,height=768,res=150)
manhattan(subset(data, CHR==9), main = "Chromosome 9", ylim = c(0, 7), cex = 0.6, 
         cex.axis = 1, cex.lab = 1, cex.main = 1, col = c("blue4", "red3"), genomewideline = F,
         annotatePval = 0.00001, annotateTop = FALSE)
dev.off()
```
